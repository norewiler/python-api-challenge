```python
# Import Dependencies
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import requests
import gmaps
import os
import json
from scipy.stats import linregress
from citipy import citipy
from datetime import date

# Import API key
from api_keys import weather_api_key
```

## Generate Cities List

```python
cities = []
i = 0
while i <575:
    lat = np.random.uniform(-90, 90)
    lng = np.random.uniform(-180, 180)
    city = citipy.nearest_city(lat, lng).city_name
    if city not in cities:
        cities.append(city)
        i += 1
print(len(cities))
        

```

### Perform API Calls
* Perform a weather check on each city using a series of successive API calls.
* Include a print log of each city as it'sbeing processed (with the city number and city name).


```python
#build URL
base_url = "http://api.openweathermap.org/data/2.5/forecast?id=524901&APPID="
my_url = base_url + weather_api_key

# samp_city = cities[0]
# samp_url = my_url + "&q=" + samp_city + "&units=Imperial"
# print(samp_url)
# response = requests.get(samp_url)
# json_resp = response.json()
# print(type(json_resp["city"]["name"]))
# # print(json.dumps(json_resp, indent=4))

valid_count = 0
city_names = []
lats = []
lngs = []
temps = []
humid = []
cloudiness = []
wind = []
country = []
date = []

for city in cities:
    #print(city)
    req_url = my_url + "&q=" + city + "&units=Imperial"
    resp = requests.get(req_url)
    #print(resp)
    j_resp = resp.json()
    try:
        city_names.append(j_resp["city"]["name"])
        lats.append(j_resp["city"]["coord"]["lat"])
        lngs.append(j_resp["city"]["coord"]["lon"])
        temps.append(j_resp["list"][0]["main"]["temp_max"])
        humid.append(j_resp["list"][0]["main"]["humidity"])
        cloudiness.append(j_resp["list"][0]["clouds"]["all"])
        wind.append(j_resp["list"][0]["wind"]["speed"])
        country.append(j_resp["city"]["country"])
        date.append(j_resp["list"][0]["dt"])
        valid_count += 1
        print(f"City {valid_count}: {city}")
    except:
        print(f"Request for {city} generated a {resp} response.") 
        pass
    
print(len(city_names))

```

### Convert Raw Data to DataFrame
* Export the city data into a .csv.
* Display the DataFrame

```python
weather_df = pd.DataFrame({
    "City": city_names,
    "Lat": lats,
    "Lng": lngs,
    "Max Temp": temps,
    "Humidity": humid,
    "Cloudiness" : cloudiness,
    "Wind Speed": wind,
    "Country": country,
    "Date": date
})
weather_df.to_csv("outputData/weather.csv")
weather_df.head()
```

## Linear Regression

```python
northern_data = clean_city_data[clean_city_data["Lat"] > 0]
n_latPoints = northern_data["Lat"]
southern_data = clean_city_data[clean_city_data["Lat"] < 0]
s_latPoints = southern_data["Lat"]
print(clean_city_data.shape, northern_data.shape, southern_data.shape)
```

####  Northern Hemisphere - Max Temp vs. Latitude Linear Regression

```python
n_tempPoints = northern_data["Max Temp"]
# Run the regression and store each of the returned values
slope, intercept, rvalue, pvalue, stderr = linregress(n_latPoints, n_tempPoints)

# Calculate the predictions for each value of x (# Rooms)
regression_values = n_latPoints * slope + intercept
plt.scatter(n_latPoints,n_tempPoints)
plt.plot(n_latPoints, regression_values, "red")
plt.title('Northern Hemisphere: City Latitude vs. Max Temperature')
plt.xlabel('Latitude')
plt.ylabel('Max Temperature (F)')
plt.xlim(-10,90)
plt.ylim(0,120)
print(f"The r-value is: {rvalue}.")
```

####  Southern Hemisphere - Max Temp vs. Latitude Linear Regression

```python
s_tempPoints = southern_data["Max Temp"]
# Run the regression and store each of the returned values
slope, intercept, rvalue, pvalue, stderr = linregress(s_latPoints, s_tempPoints)

# Calculate the predictions for each value of x (# Rooms)
regression_values = s_latPoints * slope + intercept
plt.scatter(s_latPoints,s_tempPoints)
plt.plot(s_latPoints, regression_values, "red")
plt.title('Southern Hemisphere: City Latitude vs. Max Temperature')
plt.xlabel('Latitude')
plt.ylabel('Max Temperature (F)')
plt.xlim(-90,10)
plt.ylim(0,120)
print(f"The r-value is: {rvalue}.")
```

Both regressions show that as distance from the equator increases, temperature decreases.
