```python
# Import Dependencies
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import requests
import gmaps
import os
import json

from citipy import citipy

# Import API key
from api_keys import weather_api_key
```

## Generate Cities List

```python
cities = []
i = 0
while i <575:
    lat = np.random.uniform(-90, 90)
    lng = np.random.uniform(-180, 180)
    city = citipy.nearest_city(lat, lng).city_name
    if city not in cities:
        cities.append(city)
        i += 1
print(len(cities))
        

```

### Perform API Calls
* Perform a weather check on each city using a series of successive API calls.
* Include a print log of each city as it'sbeing processed (with the city number and city name).


```python
#build URL
base_url = "http://api.openweathermap.org/data/2.5/forecast?id=524901&APPID="
my_url = base_url + weather_api_key

# samp_city = cities[0]
# samp_url = my_url + "&q=" + samp_city + "&units=Imperial"
# print(samp_url)
# response = requests.get(samp_url)
# json_resp = response.json()
# print(type(json_resp["city"]["name"]))
# # print(json.dumps(json_resp, indent=4))

valid_count = 0
city_names = []
lats = []
lngs = []
temps = []
humid = []
cloudiness = []
wind = []
country = []
date = []

for city in cities:
    #print(city)
    req_url = my_url + "&q=" + city + "&units=Imperial"
    resp = requests.get(req_url)
    #print(resp)
    j_resp = resp.json()
    try:
        city_names.append(j_resp["city"]["name"])
        lats.append(j_resp["city"]["coord"]["lat"])
        lngs.append(j_resp["city"]["coord"]["lon"])
        temps.append(j_resp["list"][0]["main"]["temp_max"])
        humid.append(j_resp["list"][0]["main"]["humidity"])
        cloudiness.append(j_resp["list"][0]["clouds"]["all"])
        wind.append(j_resp["list"][0]["wind"]["speed"])
        country.append(j_resp["city"]["country"])
        date.append(j_resp["list"][0]["dt"])
        valid_count += 1
        print(f"City {valid_count}: {city}")
    except:
        print(f"Request for {city} generated a {resp} response.") 
        pass
    
print(len(city_names))

```

### Convert Raw Data to DataFrame
* Export the city data into a .csv.
* Display the DataFrame

```python
weather_df = pd.DataFrame({
    "City": city_names,
    "Lat": lats,
    "Lng": lngs,
    "Max Temp": temps,
    "Humidity": humid,
    "Cloudiness" : cloudiness,
    "Wind Speed": wind,
    "Country": country,
    "Date": date
})
weather_df.to_csv("outputData/weather.csv")
weather_df.head()
```

## Inspect the data and remove the cities where the humidity > 100%.

## Plotting the Data

## Latitude vs. Temperature Plot

```python
latPoints = clean_city_data["Lat"]
tempPoints = clean_city_data["Max Temp"]
plt.scatter(latPoints,tempPoints)
plt.title('City Latitude vs. Max Temperature')
plt.xlabel('Latitude')
plt.ylabel('Max Temperature (F)')
plt.xlim(-90,90)
plt.ylim(0,120)
plt.savefig('outputData/lat_vs_temp.png')

```

This plot shows that temperature rises as you get closer to the equator. (Absolute value of Latitiude is closer to 0)

## Latitude vs. Humidity Plot

```python
humidPoints = clean_city_data["Humidity"]
plt.scatter(latPoints,humidPoints)
plt.title('City Latitude vs. Humidity')
plt.xlabel('Latitude')
plt.ylabel('Humidity (%)')
plt.xlim(-90,90)
plt.ylim(-5,105)
plt.savefig('outputData/lat_vs_humidity.png')

```

This plot suggests that countries on the northern hemisphere (lat > 0) tend to be more humid than those on the southern hemisphere. I say suggests becuse there doen't seem to be a very stong correlation.

## Latitude vs. Cloudiness Plot

```python
cloudPoints = clean_city_data["Cloudiness"]
plt.scatter(latPoints,cloudPoints)
plt.title('City Latitude vs. Cloudiness')
plt.xlabel('Latitude')
plt.ylabel('Cloudiness (%)')
plt.xlim(-90,90)
plt.ylim(-5,105)
plt.savefig('outputData/lat_vs_cloudiness.png')
```

This plot doesn't show much of a correlation between latitude and cloudiness.

## Latitude vs. Wind Speed Plot

```python
windPoints = clean_city_data["Wind Speed"]
plt.scatter(latPoints,windPoints)
plt.title('City Latitude vs. Wind Speed')
plt.xlabel('Latitude')
plt.ylabel('Wind Speed (mph)')
plt.xlim(-90,90)

plt.savefig('outputData/lat_vs_wind.png')
```

This plot analyzes the relationship between latitude and wind speed and seems to suggest that the relationship is weak.
